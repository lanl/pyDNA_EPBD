```python
import os
import sys
home_dir = "../"
module_path = os.path.abspath(os.path.join(home_dir))
if module_path not in sys.path:
    sys.path.append(module_path)

import utils.pickle_utils as pickle_utils
from utils.dicts import colors_dict
import numpy as np
import matplotlib.pyplot as plt
```

```python
root = "/usr/projects/pyDNA_EPBD/pyDNA_EPBD_inputs_outputs/"
simulation_inps_dir = root+"data/88_sequences/"
simulation_outs_dir = root+"outputs_simulation/88_seqs/88_sequences/"
```

```python
n_steps_after_preheating = 80000
norm = 1.00 / (n_steps_after_preheating)
```

```python
def plot_features(features, seq_name, color="gray", y_label="Average coordinate distance", out_dir="coord_profiles"):
    out_dir = home_dir+f"plots/{out_dir}/"
    os.makedirs(out_dir, exist_ok=True)
    
    plt.clf()
    features = features*norm
    plt.plot(features, color=color, marker=".", linestyle='--', markersize=5, linewidth=.5, markerfacecolor='none')
    plt.xlabel("DNA Nucleotide index")
    plt.ylabel(y_label)
    plt.title(seq_name)
    # plt.show()
    plt.savefig(out_dir+f"{seq_name}.png", dpi=300, format="png", bbox_inches='tight', pad_inches=0.05)
```

```python
with open(simulation_inps_dir+"88_sequences_ids.txt") as f:
    for line in f.readlines():
        
        seq_id, seq_name, seq_len = line.split(":")
        features = pickle_utils.load_pickle(simulation_outs_dir+f"{seq_id}.pkl")
        print(seq_name)
        for key, value in features.items():
            print("\t", key, features[key].shape)

        plot_features(features["flip_verbose"][:, 1],  seq_name, color=colors_dict["flips"], y_label="Average flipping profile", out_dir="88seqs_flip_profiles") # th 1.414213562373096
        # plot_features(features["coord"], seq_name, color=colors_dict["coords"], y_label="Average coordinate distance", out_dir="88seqs_coord_profiles")

        # break
```
