```python
import os
import sys
home_dir = "../"
module_path = os.path.abspath(os.path.join(home_dir))
if module_path not in sys.path:
    sys.path.append(module_path)

import pandas as pd
from models.helpers import data_format
```

```python
datatype = "selex" # gcpbm, selex
root = "/usr/projects/pyDNA_EPBD/pyDNA_EPBD_inputs_outputs/"
input_seqs_dir = root+"data/selexdata_sequence/"
flank_size = 26
ml_outputs_dir = root+f"outputs_mls/"
print(input_seqs_dir, ml_outputs_dir, sep="\n")
```

```python
df = pd.read_csv(f"{ml_outputs_dir}aggregated_performance_{datatype}.tsv", sep="\t")
df
```

```python
seq_files_list = sorted(os.listdir(input_seqs_dir))

for seq_filename_with_ext in seq_files_list:
    seq_filename_wo_ext = seq_filename_with_ext[:-4]

    x = df.loc[df["Data"]==seq_filename_wo_ext]
    if x.shape[0] == 0: 
        print(seq_filename_wo_ext)
        continue
    
    loci = x.index[0]
    start_loci, end_loci = loci, loci+4

    df.loc[start_loci:end_loci, "Data"] = seq_filename_wo_ext
```

```python
rbf_svr_df = df[df["Method"]=="RBF_SVR"]
rbf_svr_df.drop(columns=["Method"], inplace=True)
rbf_svr_df
```

```python
onehot_r2 = rbf_svr_df["Onehot R2(avg/std)"].apply(lambda x: float(x.split("/")[0])).to_numpy()
onehotbds_r2 = rbf_svr_df["Onehot+BDs R2(avg/std)"].apply(lambda x: float(x.split("/")[0])).to_numpy()
families = rbf_svr_df["Data"].apply(lambda x: x.split("_")[0]).to_numpy()
print(onehot_r2.shape, onehotbds_r2.shape, families.shape)
print(len(set(families)))
```

```python
rbf_svr_df.to_csv(f"{ml_outputs_dir}aggregated_performance_{datatype}_rbf_svr.tsv", index=False, sep="\t")
```
